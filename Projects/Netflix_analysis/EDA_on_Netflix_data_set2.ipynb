<p>What is the year that has highest number of movie releases?<p>
<p>show each year movie realease in a bar chart?<p>

```python
new_df = df.dropna()
new_df.isnull().sum()  # new data frame from the original one that has no null values
```

```python
new_df['new_date'] = pd.to_datetime(new_df['Release_Date'], format='mixed') 
```

```python
new_df.head()
```

```python
#Number of content released on each year
new_df['year_count'] = new_df.new_date.dt.year
new_df
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=new_df, x = 'year_count', order=new_df.new_date.dt.year.value_counts().index[0:10])
```

<p>Using bar plot<p>

```python
new_df.new_date.dt.year.value_counts().plot(kind='bar')
```

<p>How many Movies & Tv shows are in the dataset? Show with bar graph.<p>

```python
#Using value count mehtod
df.Category.value_counts()
```

```python
#using gropuing techique
df.groupby('Category').Category.count()
```
